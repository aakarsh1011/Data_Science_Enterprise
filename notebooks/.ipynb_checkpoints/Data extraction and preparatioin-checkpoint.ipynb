## Importing necessary libraries

```python
import pandas as pd      # for data manupulations
import matplotlib.pyplot as plt # for visualizations
import missingno as msno
```

```python
# url from the Covid-19 data is extracted
url = "https://covid.ourworldindata.org/data/owid-covid-data.csv"
```

```python
covid_data = pd.read_csv(url,sep=",")
```

```python
# looking at the shape and other properties of the data

print(f"We have {covid_data.shape[0]} rows and {covid_data.shape[1]} columns")
```

```python
print(f"The number of unique countries in the data are {len(covid_data['location'].unique())}")
```

##### We will be taking 3 countries to visualize the results

- India
- Germany
- France

```python
countries = ["India", "Germany","France"]
countries_df = covid_data[covid_data["location"].isin(countries)]
```

Let us take a look at the sample data now

```python
print(f"We have {countries_df.shape[0]} rows and {countries_df.shape[1]} columns")
print(f"The number of unique countries in the data are {len(countries_df['location'].unique())}")
```

```python
# Lets check how many null values we have in our required data
msno.matrix(countries_df[['location','date','total_cases','total_vaccinations','population']])
```

Most of the data is available except the vaccinations, let us see with respect to each country

```python
df_Ind = covid_data[covid_data['location'] == "India"]
df_Ger = covid_data[covid_data['location'] == "Germany"]
df_Fra = covid_data[covid_data['location'] == "France"]
```

```python
fig, ax = plt.subplots(figsize=(20, 5),nrows=1,ncols=3)
msno.matrix(df_Ind[['location','date','total_cases','total_vaccinations','population']],ax=ax[0], sparkline=False)
msno.matrix(df_Ger[['location','date','total_cases','total_vaccinations','population']],ax=ax[1], sparkline=False)
msno.matrix(df_Fra[['location','date','total_cases','total_vaccinations','population']],ax=ax[2], sparkline=False)
```

Vaccination seems consitent, except for initial days maybe when it was not available

For the first task, we need the ratio of the total cases and population.
So lets create a seperate column for the same

```python
countries_df['cases2population'] = countries_df.apply(lambda x: x['total_cases']/x['population'], axis=1)
```

```python
msno.matrix(countries_df[['location','date','total_cases','total_vaccinations','population','cases2population']])
```

Now we can save the dataset for further use

```python
countries_df = countries_df.reset_index(drop=True)
countries_df
```
