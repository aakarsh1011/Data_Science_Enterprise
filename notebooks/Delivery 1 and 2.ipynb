```python
# importing neccesary libraries for the visualization
import pandas as pd
import plotly.express as px
%matplotlib inline
```

```python
# importing the data of countries
# India, Germany, France
countries_df = pd.read_csv("E:/uni/Data_Science_E/Data_Science_Enterprise/data/countries_df.csv")
countries_df.head(5)
```

## Delivery 1
The relative cases overtime of Covid infectors (absolut Covid cases/population size)

```python
fig = px.line(countries_df, x='date',y='cases2population',color='location',title = "COVID cases")
fig.show()
```

## Delivery 2
The vaccination rate (percentage of the population) over time

for vaccination percentage, we will have to compute

vaccination rate = (total_vaccinations/population)*100

```python
countries_df['vaccination_rate'] = countries_df.apply(lambda x: (x['total_cases']/x['population'])*100, axis=1)
```

```python
fig = px.line(countries_df, x='date',y='vaccination_rate',color='location',title = "Vaccination Rate")
fig.show()
```
